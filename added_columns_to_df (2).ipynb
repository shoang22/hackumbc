```
import pandas as pd 
import numpy as np 
```

```
#import the data 
import requests
fake_news_file = "https://raw.githubusercontent.com/cuilimeng/CoAID/master/09-01-2020/NewsFakeCOVID-19.csv"
res = requests.get(fake_news_file, allow_redirects=True)
with open('NewsFakeCOVID-19.csv','wb') as file:
    file.write(res.content)
fake_covid_news = pd.read_csv('NewsFakeCOVID-19.csv')
```

```
#Read the file into a dataframe
df = pd.read_csv('NewsFakeCOVID-19.csv')

#show the data 
df['news_url']
```

```
#Create new columns for just the news source
df['news_source']= df["news_url"].str.split('/').str.get(2)
df[['news_url', 'news_source']]

```
