```python
import pandas as pd

import glob

# read all csv's in DATA

#path = r'C:/Users/15714/Documents/GMU/Classes/senior/CDS-490/DATA' # use your path
all_files = glob.glob("./ds2/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
df = pd.read_csv('./ds1/FakeCovid_June2020.csv')
```

```python
df.head()
```

```python
df1 = frame[['news_url','title','content','fake']]
```

```python
df1.head()
```

```python
df2 = df[['article_source','content_text','fake','lang','source_title']]
df2 = df2[df2['lang']=='English']
```

```python
df2.rename(columns={"article_source": "news_url", "source_title": "title",'content_text':'content'},inplace=True)
```

```python
dff = pd.concat([df1,df2], axis=0, ignore_index=True)
```

```python
dff.drop('lang',axis=1,inplace=True)
```

```python
dff
```

```python
#Create new columns for just the news source
dff['news_source']= dff["news_url"].str.split('/').str.get(2)
```

```python
dff['news_source'] = dff['news_source'].map(lambda x: str(x).lstrip('w.'))
```

```python
dff['combined_text'] = dff['title'] + dff['content'] + dff['news_source'].astype(str)
```
